## Carregando os Dados

### Arquivo JSON

```python
data = pd.read_json(path_or_buf='imoveis.json', orient='columns')
data
```

### Arquivo Excel

```python
bairros = pd.read_excel(
    io='bairros.xlsx',
    sheet_name='Preço médio por tipo', #Nome da página dentro do arquivo
    usecols='C:E', #Quais colunas pegar
    header=2, #Em qual linha começar
    index_col=[0,1], #usar as 2 primeiras colunas como índice,
    names=['bairros', 'tipos', 'valor_m2_bairro']
)
bairros
```

## Trasformando Dados JSON Para uma Tabela

```python
data_normal = pd.json_normalize(data=data['normal']) #Aqui nós pegamos a "primeira" coluna
data_normal
```

```python
#Podemos observar que é uma lista de arquivos Json
data_normal['listings'].iloc[0]
```

```python
#Para confirmar:
type(data_normal['listings'].iloc[0])
```

```python
#Vamos ver quantos elementos existem dentro dessa lista
len(data_normal['listings'].iloc[0])
```

Vemos que existem 50 arquivos em formato de dicionários, então agora vamos começar com a normalização dos dados

```python
data_normal_listings = pd.json_normalize(data=data_normal['listings'].iloc[0], sep='_')
data_normal_listings
```

```python
#Se observarmos, vamos ver que na coluna "Highlights" também temos "Listings"
data['highlights'].iloc[0]
```

```python
#Então vamos fazer a mesma coisa pra coluna "Highlights"
data_highlights_listings = pd.json_normalize(data=data['highlights'], sep='_', record_path=['listings'])
data_highlights_listings
```

### Transformando Strings em Listas

```python
data_normal_listings.head(2)
```

Se observarmos as colunas 7, 8, e 9 veremos que elas aparentam ser Listas. Porém na verdade elas são do tipo String, vamos ver:

```python
#Aqui nós estamos pegando só a primeira linha da coluna especificada
lista_str = data_normal_listings.loc[0, 'imovel_caracteristicas_propriedade']
lista_str
```

```python
type(lista_str)
```

```python
#Vamos acessar o primeiro elemento para ver qual é
lista_str[0]
```

E foi exatamento como o esperado, pelo fato de não ser uma lista ele nos retorna apenas um caracter ao invés de o elemento "Área de Serviço".


Após essas demonstrações, agora vamos fazer nossas alterações de String para Lista.

```python
colunas = data_normal_listings.filter(like='imovel_caracteristicas').columns

for col in colunas:
    data_normal_listings[col] = data_normal_listings[col].astype(str) \
    .str[1:-1].str.replace("'", '').str.split(", ")
    
    data_highlights_listings[col] = data_highlights_listings[col].astype(str) \
    .str[1:-1].str.replace("'", "").str.split(", ")
```

```python
tipo_lista = data_normal_listings.loc[0, 'imovel_caracteristicas_propriedade']
tipo_lista
```

```python
type(tipo_lista)
```

```python
tipo_lista[0]
```

Transformação feita

### Empilhando DataFrames

```python
data_listings = pd.concat([data_normal_listings, data_highlights_listings], ignore_index=True)
data_listings.head()
```

### Combinando os *DataFrames*

Vamos combinar nosso DataFrame(Json) com o outro DataFrame(Excel), utilizando a função Merge

```python
data_listings = pd.merge(
    left = data_listings,
    right = bairros,
    left_on = ['imovel_endereco_bairro', 'imovel_tipos_propriedade'],
    right_index = True
)
data_listings.head()
```

---
## Criando as colunas "quartos", "suites" e "banheiros"

```python
data_listings['anuncio_descricao'].values[:10]
```

```python
configuracao = data_listings['anuncio_descricao'].str.extractall('(\d+)')
configuracao.head(9)
```

```python
#Transformando em tabela
configuracao = configuracao.unstack()
configuracao = configuracao.rename(columns={0:'quartos', 1:'suites', 2:'banheiros'})
configuracao
```

Vamos remover agora esse "segundo" nome da coluna quartos

```python
configuracao = configuracao.droplevel(level=0, axis=1)
configuracao
```

```python
data_listings = pd.merge(data_listings, configuracao, left_index=True, right_index=True)
data_listings
```

---
## Criando novas classificações

```python
tipo = {
    'Apartamento': 'Residencial',
    'Casa': 'Residencial',
    'Cobertura': 'Residencial',
    'Consultório': 'Comercial',
    'Imóvel Comercial': 'Comercial',
    'Loja': 'Comercial',
    'Sala Comercial': 'Comercial'
}

data_listings['imovel_tipos_propriedade'].map(tipo)
```

Criamos uma classificação para ficar mais fácil de identificar qual é o tipo de imóvel, agora falta inserir ele no nosso DataFrame

```python
imovel_tipos = data_listings['imovel_tipos_propriedade'].map(tipo)
```

```python
data_listings.insert(loc=0, column='imovel_tipos_uso', value=imovel_tipos)
```

```python
data_listings.head()
```

### Criando Classes

Vamos criar algumas classes, para identificar se o imóvel é "Popular", "Padrão" ou "Alto Padrão"

```python
valor_minimo = data_listings['anuncio_valores_venda'].min()

valor_maximo = data_listings['anuncio_valores_venda'].max()

rotulos = ['Popular', 'Padrão', 'Alto Padrão']

classes = [valor_minimo, 400000, 2000000, valor_maximo]
```

```python
pd.cut(x=data_listings['anuncio_valores_venda'], bins=classes, labels=rotulos, include_lowest=True)
```

```python
data_listings['classe_valor'] = pd.cut(x=data_listings['anuncio_valores_venda'], bins=classes, labels=rotulos, include_lowest=True)
data_listings.head()
```

## Criando Novas Colunas Apartir das Existentes

### Valor por m²

```python
data_listings['valor_m2'] = data_listings['anuncio_valores_venda'] / data_listings['imovel_area']

data_listings[['anuncio_valores_venda', 'imovel_area', 'valor_m2']].iloc[16:19]
```

Se observarmos, na coluna valor_m2 temos um "inf". Então vamos ter que fazer um tratamento para que esse "inf" se torne um 0

```python
valor_m2 = lambda dado: dado['anuncio_valores_venda'] / dado['imovel_area'] if dado['imovel_area'] != 0 else 0

data_listings['valor_m2'] = data_listings.apply(valor_m2, axis=1)

data_listings[['anuncio_valores_venda', 'imovel_area', 'valor_m2']].iloc[16:19]
```

### Tem Piscina?

```python
data_listings['Piscina_1'] = data_listings['imovel_caracteristicas_condominio'].apply(lambda x: 'Piscina' in x)
data_listings[['imovel_caracteristicas_condominio', 'Piscina_1']].head()
```

```python
sim_nao = {
    False: "Não",
    True: "Sim"
}

sim_nao = data_listings['Piscina_1'].map(sim_nao)

data_listings.insert(loc=24,column="Piscina", value=sim_nao)
```

```python
data_listings.drop(columns="Piscina_1", inplace=True)
```
